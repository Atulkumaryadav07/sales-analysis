## Extract Archive

### Subtask:
Extract the contents of the `/content/archive.zip` file to access the sales data.


**Reasoning**:
To extract the contents of the `/content/archive.zip` file, I need to use the `zipfile` module to handle the archive and the `os` module to create the destination directory.



```python
import zipfile
import os

zip_file_path = '/content/archive.zip'
extraction_dir = 'extracted_data'

# Create the extraction directory if it doesn't exist
os.makedirs(extraction_dir, exist_ok=True)

# Open the zip file and extract its contents
with zipfile.ZipFile(zip_file_path, 'r') as zip_ref:
    zip_ref.extractall(extraction_dir)

print(f"Archive '{zip_file_path}' extracted to '{extraction_dir}'")
print(f"Contents of '{extraction_dir}':\n{os.listdir(extraction_dir)}")
```

## Identify and Load Sales Data

### Subtask:
Identify the sales data file (e.g., CSV, Excel) from the extracted contents and load it into a pandas DataFrame.


**Reasoning**:
To identify and load the sales data, I need to import the pandas library, construct the full path to the `AI_csv.csv` file within the `extracted_data` directory, and then use `pd.read_csv` to load it into a DataFrame named `df_sales`.



```python
import pandas as pd

sales_file_path = os.path.join(extraction_dir, 'AI_csv.csv')

df_sales = pd.read_csv(sales_file_path)

print(f"Sales data loaded successfully from: {sales_file_path}")
print("First 5 rows of the sales data:")
df_sales.head()
```

## Analyze Key Sales Metrics

### Subtask:
Analyze key attendance metrics, adapting from the original sales metrics. This involves cleaning the DataFrame header and data, converting data types, and calculating relevant attendance statistics for each student.


**Reasoning**:
To begin the data cleaning process, I will define the new, more descriptive column names for the `df_sales` DataFrame, incorporate the sub-header information from the first row, apply these names, remove the now redundant first row, and then reset the DataFrame's index to ensure a clean structure for further analysis.



```python
new_column_names = [
    'Roll No',
    'Student Name',
    '09-Jan',
    '10-Jan',
    '16-Jan',
    '17-Jan',
    '23-Jan',
    '24-Jan',
    '30-Jan',
    '31-Jan',
    'Current Month Attended Lectures', # Curent Month's Lect. Att.
    'Current Month Delivered Lectures', # Unnamed: 11 Del.
    'Current Month Attendance Percentage', # Unnamed: 12 %
    'Overall Attended Lectures', # Overall Lect. Att.
    'Overall Delivered Lectures', # Unnamed: 14 Del.
    'Overall Attendance Percentage' # Unnamed: 15 %
]

df_sales.columns = new_column_names

# Remove the first row which contains sub-headers and NaN values
df_sales = df_sales.drop(0).reset_index(drop=True)

print("DataFrame after cleaning header and removing first row:")
print(df_sales.head())
print("\nNew Column Names:")
print(df_sales.columns)
```

**Reasoning**:
To continue cleaning the data, I will convert the 'Roll No' column to integer, convert date columns from 'P'/'A' to 1/0 and then to integer, convert lecture count columns to integer, and clean and convert percentage columns to float. Afterwards, I will calculate 'Total Daily Presents', 'Calculated Current Month Attendance Percentage', and 'Calculated Overall Attendance Percentage' based on the cleaned data, and then display the first 10 rows of the updated DataFrame.



```python
df_sales['Roll No'] = df_sales['Roll No'].astype(int)

date_columns = ['09-Jan', '10-Jan', '16-Jan', '17-Jan', '23-Jan', '24-Jan', '30-Jan', '31-Jan']
for col in date_columns:
    df_sales[col] = df_sales[col].replace({'P': 1, 'A': 0, '': 0}).fillna(0).astype(int)

lecture_count_columns = [
    'Current Month Attended Lectures',
    'Current Month Delivered Lectures',
    'Overall Attended Lectures',
    'Overall Delivered Lectures'
]
for col in lecture_count_columns:
    df_sales[col] = pd.to_numeric(df_sales[col], errors='coerce').fillna(0).astype(int)

percentage_columns = [
    'Current Month Attendance Percentage',
    'Overall Attendance Percentage'
]
for col in percentage_columns:
    df_sales[col] = df_sales[col].astype(str).str.replace('%', '').astype(float)

# Calculate Total Daily Presents
df_sales['Calculated Total Daily Presents'] = df_sales[date_columns].sum(axis=1)

# Calculate Current Month Attendance Percentage
df_sales['Calculated Current Month Attendance Percentage'] = (
    df_sales['Current Month Attended Lectures'] / df_sales['Current Month Delivered Lectures']
) * 100
df_sales['Calculated Current Month Attendance Percentage'] = df_sales['Calculated Current Month Attendance Percentage'].fillna(0)

# Calculate Overall Attendance Percentage
df_sales['Calculated Overall Attendance Percentage'] = (
    df_sales['Overall Attended Lectures'] / df_sales['Overall Delivered Lectures']
) * 100
df_sales['Calculated Overall Attendance Percentage'] = df_sales['Calculated Overall Attendance Percentage'].fillna(0)

print("DataFrame after data type conversions and metric calculations (first 10 rows):")
print(df_sales.head(10))
```

**Reasoning**:
The previous code failed because there were literal space characters (' ') in the date columns that were not handled by the `replace` method, leading to a `ValueError` when attempting to convert to integer. I will modify the `replace` dictionary to also convert these space characters to 0 before casting to integer.



```python
df_sales['Roll No'] = df_sales['Roll No'].astype(int)

date_columns = ['09-Jan', '10-Jan', '16-Jan', '17-Jan', '23-Jan', '24-Jan', '30-Jan', '31-Jan']
for col in date_columns:
    df_sales[col] = df_sales[col].replace({'P': 1, 'A': 0, '': 0, ' ': 0}).fillna(0).astype(int)

lecture_count_columns = [
    'Current Month Attended Lectures',
    'Current Month Delivered Lectures',
    'Overall Attended Lectures',
    'Overall Delivered Lectures'
]
for col in lecture_count_columns:
    df_sales[col] = pd.to_numeric(df_sales[col], errors='coerce').fillna(0).astype(int)

percentage_columns = [
    'Current Month Attendance Percentage',
    'Overall Attendance Percentage'
]
for col in percentage_columns:
    df_sales[col] = df_sales[col].astype(str).str.replace('%', '').astype(float)

# Calculate Total Daily Presents
df_sales['Calculated Total Daily Presents'] = df_sales[date_columns].sum(axis=1)

# Calculate Current Month Attendance Percentage
df_sales['Calculated Current Month Attendance Percentage'] = (
    df_sales['Current Month Attended Lectures'] / df_sales['Current Month Delivered Lectures']
) * 100
df_sales['Calculated Current Month Attendance Percentage'] = df_sales['Calculated Current Month Attendance Percentage'].fillna(0)

# Calculate Overall Attendance Percentage
df_sales['Calculated Overall Attendance Percentage'] = (
    df_sales['Overall Attended Lectures'] / df_sales['Overall Delivered Lectures']
) * 100
df_sales['Calculated Overall Attendance Percentage'] = df_sales['Calculated Overall Attendance Percentage'].fillna(0)

print("DataFrame after data type conversions and metric calculations (first 10 rows):")
print(df_sales.head(10))
```

**Reasoning**:
The previous steps successfully cleaned the data and calculated key attendance metrics. Now, to analyze the overall attendance, I will calculate and display descriptive statistics for the newly created calculated attendance percentage columns, identify the top and bottom 5 students by overall attendance, and also determine the average daily attendance.



```python
print("\nDescriptive Statistics for Attendance Percentages:")
print(df_sales[['Calculated Current Month Attendance Percentage', 'Calculated Overall Attendance Percentage']].describe())

print("\nAverage daily presents:")
print(df_sales[date_columns].sum().mean())

print("\nStudents with Highest Overall Attendance:")
print(df_sales.sort_values(by='Calculated Overall Attendance Percentage', ascending=False).head()[['Student Name', 'Calculated Overall Attendance Percentage']])

print("\nStudents with Lowest Overall Attendance:")
print(df_sales.sort_values(by='Calculated Overall Attendance Percentage', ascending=True).head()[['Student Name', 'Calculated Overall Attendance Percentage']])
```

## Visualize Sales Trends

### Subtask:
Generate plots to visualize the attendance data, focusing on trends and distribution, and include appropriate legends and labels.


**Reasoning**:
To visualize the attendance data, I will import `matplotlib.pyplot` and `seaborn`, then create a histogram for the distribution of 'Calculated Overall Attendance Percentage' and a bar chart for the daily attendance trends, ensuring proper titles and labels for both plots.



```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set style for plots
sns.set_style("whitegrid")

# 1. Histogram for Distribution of Overall Attendance Percentage
plt.figure(figsize=(10, 6))
sns.histplot(df_sales['Calculated Overall Attendance Percentage'], bins=10, kde=True)
plt.title('Distribution of Overall Attendance Percentage')
plt.xlabel('Overall Attendance Percentage')
plt.ylabel('Number of Students')
plt.show()

# 2. Bar chart for Daily Attendance Trend
daily_presents = df_sales[date_columns].sum()
plt.figure(figsize=(12, 6))
sns.barplot(x=daily_presents.index, y=daily_presents.values, palette='viridis')
plt.title('Daily Attendance Trend')
plt.xlabel('Date')
plt.ylabel('Total Presents')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

**Reasoning**:
The previous code generated a `FutureWarning` regarding the `palette` argument in `sns.barplot`. To address this and align with future seaborn versions, I will explicitly set the `hue` argument to `daily_presents.index` and `legend` to `False` in the bar plot.



```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set style for plots
sns.set_style("whitegrid")

# 1. Histogram for Distribution of Overall Attendance Percentage
plt.figure(figsize=(10, 6))
sns.histplot(df_sales['Calculated Overall Attendance Percentage'], bins=10, kde=True)
plt.title('Distribution of Overall Attendance Percentage')
plt.xlabel('Overall Attendance Percentage')
plt.ylabel('Number of Students')
plt.show()

# 2. Bar chart for Daily Attendance Trend
daily_presents = df_sales[date_columns].sum()
plt.figure(figsize=(12, 6))
sns.barplot(x=daily_presents.index, y=daily_presents.values, hue=daily_presents.index, palette='viridis', legend=False)
plt.title('Daily Attendance Trend')
plt.xlabel('Date')
plt.ylabel('Total Presents')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
